```python
# What is multithreading in python? why is it used? Name the module used to handle threads in python.
"""
In Python, multithreading refers to the concurrent execution of multiple threads within a single program. 
A thread is a lightweight process that can perform tasks independently while sharing the same memory space as other threads. 
"""
"""
Multithreading is used to achieve parallelism or concurrent execution, enabling programs to efficiently utilize system 
resources and improve performance.
"""
"""
In Python, the module commonly used to handle threads is called threading. 
The threading module provides a high-level interface for creating and managing threads in Python. 
It allows you to create new threads, start them, control their execution, and coordinate their interaction.
"""
```

```python
""" 
Why threading module used? write the use of the following functions
1. activeCount
2. currentThread
3. enumerate
"""

"""
The threading module in Python is used for creating and managing threads.
It provides a way to run multiple threads concurrently, allowing for better utilization of system resources and enabling concurrent
execution of tasks
"""

"""
1. activeCount
The activeCount() function returns the number of Thread objects currently alive.
It is a class method of the threading module and does not require an instance of the Thread class.
"""

"""
2. currentThread
The currentThread() function returns the current Thread object corresponding to the caller's thread of control.
It can be called from within a thread to obtain a reference to the Thread object representing that thread.
"""

"""
3. enumerate
The enumerate() function returns a list of all Thread objects currently alive.
It is a class method of the threading module and does not require an instance of the Thread class.
"""
```

```python
"""
3. Explain the following functions
1. run()
2. start()
3. join()
4. isAlive()
"""

"""
1. run()
The run() function is the entry point for a thread. It is the function that is executed when the thread is started. 
The run() function can contain any code that you want the thread to execute.

2. start()
The start() function starts a thread by calling the run() function. The start() function is a blocking function, which means that 
it will not return until the thread has finished executing the run() function.

3. join()
The join() function waits for a thread to terminate. The join() function is a blocking function, which means that it will not return 
until the thread has terminated.

4. isAlive()
The isAlive() function checks whether a thread is still executing. The isAlive() function returns True if the thread is still executing 
and False if the thread has terminated.
"""
```

```python
#Q4. Write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes

import threading

def square(n):
    print(f"square of {n} is {n**2}" )

def cube(n):
    print(f"cube of {n} is {n**3}")
    
def print_square_cube():
    thread1 = [threading.Thread(target = square, args = (i,)) for i in [1,2,3,4,5,6]]
    thread2 = [threading.Thread(target = cube, args = (i,)) for i in [1,2,3,4,5,6]]
    for t1 in thread1:
        t1.start()    
    for t2 in thread2:
        t2.start()    

print_square_cube()    
```

```python
#Q6. Explain deadlocks and race conditions.

"""
Deadlocks and race conditions are two common problems that can occur in multithreaded programs.
Deadlock is a situation in which two or more threads are blocked indefinitely, waiting for each other to release a resource.
For example, let's say there are two threads, A and B, that both need to access a file. Thread A has the file open, and then thread B tries
to open the file. Thread B will be blocked until thread A releases the file, but thread A can't release the file because it's waiting for 
thread B to finish using it. This is a deadlock.


Race condition is a situation in which the outcome of a program depends on the order in which two or more threads access a shared resource.
For example, let's say there are two threads, A and B, that are both updating the same counter.
Thread A increments the counter, and then thread B decrements the counter. If thread B starts executing before thread A, 
the counter will be decremented before it's incremented, and the final value of the counter will be wrong. This is a race condition.
"""
```

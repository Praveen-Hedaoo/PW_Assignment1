```python
#Q1. What is Flask Framework? What are the advantages of Flask Framework?

"""
Flask is a micro web framework for Python, which means it provides a simple and minimalistic approach to building web applications.
It was developed to be lightweight and easy to use while still being powerful enough to create a wide range of web applications.
"""

"""
Here are some advantages of Flask framework:

Simplicity: Flask follows a minimalistic design philosophy, making it easy to understand and get started with. 
It provides a small core with flexibility to add the features you need through extensions. 
This simplicity makes it an ideal choice for small to medium-sized projects.

Lightweight: Flask has a small codebase and minimal dependencies, resulting in a lightweight framework. 
It doesn't impose any specific way of doing things and allows developers to choose the components they need, keeping the application lean and 
efficient.

Flexibility: Flask allows developers to make their own decisions regarding the tools and libraries they want to use. 
It doesn't dictate the use of particular databases, templating engines, or other components. 
This flexibility allows developers to tailor their application to specific needs and preferences.

Extensibility: Flask is highly extensible through its rich ecosystem of extensions. These extensions provide additional functionality such as 
database integration, form validation, authentication, and more. You can choose and add only the extensions that are necessary for your application,
keeping it lightweight and focused.
"""
```

```python
#Q3. What is App routing in Flask? Why do we use app routes?

"""
In Flask, app routing refers to the process of associating URL patterns with corresponding Python functions or view handlers.
It allows you to define different routes for different parts of your web application, specifying what should happen when a particular URL is requested.
"""

"""
Here are some reasons why we use app routes in Flask:

Request Handling: App routes allow you to handle different HTTP requests (GET, POST, etc.) to specific URLs by defining appropriate view functions. 

URL Mapping: App routes help in mapping URLs to specific functions or views, providing a clear structure to your application.
It makes it easier to understand and navigate the codebase, especially in larger applications.

RESTful APIs: Flask app routes are commonly used to create RESTful APIs. Each route represents a specific API endpoint, and the associated view
function defines the behavior for that endpoint. This allows you to create APIs for retrieving, creating, updating, or deleting data.

Dynamic Routing: Flask supports dynamic routing by allowing variables in the URL patterns. 
"""
```
